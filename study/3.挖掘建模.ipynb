# 1.分类与预测
## 1.1常用的分类与预测算法
##### 回归分析
>确定预测属性（数值型）与其它变量键相互依赖的定量关系。

##### 决策树
>采用自顶向下的递归方式，在内部节点进行属性值的比较，并根据不同的属性值从该节点向下分支，最终得到的叶节点是学习划分的类。

##### 人工神经网络
>模仿大脑神经网络结构和功能而建立的信息处理系统，表示神经网络的输入与输出变量之间关系的模型。

##### 贝叶斯网络
>又称信度网络，是目前不确定知识表达和推理领域最有效的理论模型之一。

##### 支持向量机
>通过非线性映射，把低维的非线性可分转化为高维的线性可分，在高维空间进行线性分析的算法。


```python
#逻辑回归
import pandas as pd
file = r'D:\Books\Python\Python数据分析与挖掘实战\chapter5\chapter5\data\bankloan.xls'
data = pd.read_excel(file) #读入数据
data
```

```python
x = data.iloc[:,:8].as_matrix() #获取所有行前8列
y = data.iloc[:,8].as_matrix() #获取所有行第8列

import warnings
warnings.filterwarnings('ignore') #忽略运行时出现的警告信息

from sklearn.linear_model import LogisticRegression as LR
from sklearn.linear_model import RandomizedLogisticRegression as RLR
rlr = RLR() #建立随机逻辑回归模型，筛选变量
rlr.fit(x,y) #训练模型
rlr.get_support() #获得特征筛选结果
print('随机逻辑回归模型筛选特征结束！')
print('有效特征为：%s' % ','.join(data.columns[rlr.get_support()]))

x = data[data.columns[rlr.get_support()]].as_matrix() #筛选好特征

lr = LR() #建立逻辑回归模型
lr.fit(x,y) #用筛选后的特征数据来训练模型
print('逻辑回归模型训练结束！')
print('模型平均正确率为：%s' % lr.score(x,y)) #输出模型平均正确率
```

```python
#人工神经网络

import pandas as pd
file = r'D:\Books\Python\Python数据分析与挖掘实战\chapter5\chapter5\data\sales_data.xls'
data = pd.read_excel(file,index_col = '序号') #读入数据
#用1表示‘好’，‘高’，‘是’这三个属性，用0表示‘坏’，‘低’，‘否’
data[data == '好'] = 1
data[data == '是'] = 1
data[data == '高'] = 1
data[data != 1] = 0
x = data.iloc[:,:3].as_matrix().astype(int)
y = data.iloc[:,3].as_matrix().astype(int)

from keras.models import Sequential
from keras.layers.core import Dense,Activation

model = Sequential() #建立模型
model.add(Dense(input_dim=3,output_dim=10)) #添加输入层（3节点）到隐藏层（10节点）的连接
model.add(Activation('relu')) #用‘rule’函数作为激活函数，提高准确度
model.add(Dense(input_dim=10,output_dim=1))
model.add(Activation('sigmoid')) #由于是0-1输出，所以用sigmoid作为激活函数

model.compile(loss='binary_crossentropy',optimizer='adam')
#编译模型，由于是二元分类，所以指定binary_crossentropy作为损失函数

model.fit(x,y,nb_epoch=1000,batch_size=10) #训练模型，学习1000次
#绘制混淆矩阵
from sklearn.metrics import confusion_matrix #导入混淆矩阵函数
yp = model.predict_classes(x).reshape(len(y)) #分类预测
cm = confusion_matrix(y,yp) #混淆矩阵

import matplotlib.pyplot as plt
%matplotlib inline
plt.matshow(cm,cmap=plt.cm.Greens) #画混淆矩阵图，配色风格使用cm.Greens
plt.colorbar()
for i in range(len(cm)):
    for j in range(len(cm)):
        plt.annotate(cm[i,j],xy=(i,j),horizontalalignment='center',verticalalignment='center')
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()

import warnings
warnings.filterwarnings('ignore')
```

```python
#使用K-Means算法聚类消费行为特征数据
import pandas as pd
inputfile = r'D:\Books\Python\Python数据分析与挖掘实战\chapter5\chapter5\data\consumption_data.xls'
outfile = r'C:\Users\lichangcheng\Desktop\data_type.xls'
k = 3 #聚类的类别
iteration = 500 #聚类最大循环次数
data = pd.read_excel(inputfile,index_col='Id')
data_zs = 1.0*(data-data.mean())/data.std() #数据标准化
from sklearn.cluster import KMeans
model = KMeans(n_clusters=k,n_jobs=4,max_iter=iteration) #分为k类，并发数为4
model.fit(data_zs) #开始聚类
r1=pd.Series(model.labels_).value_counts() #统计各个类别的数目
r2=pd.DataFrame(model.cluster_centers_) #找出聚类中心
r=pd.concat([r2,r1],axis=1) #横向连接，得到聚类中心对应的类别下的数目，0是纵向
r.columns=list(data.columns)+['类别数目']
#print(r)
#print(data)
r=pd.concat([data,pd.Series(model.labels_,index=data.index)],axis=1)
r.columns=list(data.columns)+['聚类类别']
#print(r)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
def density_plot(data):
    import matplotlib.pyplot as plt
    plt.rcParams['font.sans-serif']=['SimHei'] #正常显示中文标签
    plt.rcParams['axes.unicode_minus'] = False #正常显示负号
    p=data.plot(kind='kde',linewidth=2,subplots=True,sharex=False)
    [p[i].set_ylabel('密度') for i in range(k)]
    plt.legend()
    return plt

for i in range(k):
    density_plot(data[r['聚类类别']==i])
    
```

#### 图表解读
> 分类1：R间隔相对较大，间隔分布在30-80天，消费次数集中在0-15次，消费金额在0-2000元。

> 分类2：R间隔在0-30天，消费次数集中在0-12次，消费金额在0-1800元。

> 分类3：R间隔在0-30天，消费次数集中在10-25次，消费金额在500-2000元。

```python
#用TSNE进行数据降维并展示聚类结果
from sklearn.manifold import TSNE
tsne = TSNE()
tsne.fit_transform(data_zs) #进行数据降维
tsne = pd.DataFrame(tsne.embedding_,index=data_zs.index) #转换数据格式
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.sans-serif']=['SimHei'] #正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False #正常显示负号
d = tsne[r['聚类类别']==0]
plt.plot(d[0],d[1],'r.')
d = tsne[r['聚类类别']==1]
plt.plot(d[0],d[1],'go')
d = tsne[r['聚类类别']==2]
plt.plot(d[0],d[1],'b*')
plt.show()
```
